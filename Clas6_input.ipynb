```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
clas6_data = pd.read_csv('clas6/CLAS6.txt', delim_whitespace=True)
clas6_data.head()
```

```python
clas6_data.drop_duplicates(keep='first')
```

```python
M_P = 0.9382720813
beam_energy = 5.75
sqrt_s = math.sqrt(2*M_P*beam_energy)
```

```python
#for factorized
file = open('clas6/clas6_grids_all.txt','w')
file.write('sqrt_s,x,z,Q,qT_Q\n')
for i,row in clas6_data.iterrows():
    x = row['x']
    z = row['z']
    Q = math.sqrt(row['Q2'])
    qT_Q = math.sqrt(row['pT2'])/z/Q
    file.write('%f,%f,%f,%f,%f\n'%(sqrt_s,x,z,Q,qT_Q))

    
```

```python
N_nodes = 36
```

```python
clas6_grids_forsmall=pd.read_csv('clas6/clas6_grids.txt')
total_line_number = len(clas6_grids_forsmall)
line_number = int(len(clas6_grids_forsmall)/(N_nodes-1))
print(line_number)
last_line_number = len(clas6_grids_forsmall)%(N_nodes-1)
print('line number for last file',last_line_number)
total_line_check = line_number*(N_nodes-1)+last_line_number
print('check origin:',total_line_number,'check after',total_line_check)
```

```python
#for factorized small files
for i_n in range(1,N_nodes):
    file = open('clas6/inputs/clas6_grids_{}.txt'.format(i_n),'w')
    #file.write('sqrt_s,x,z,Q,qT_Q\n')
    for i,row in clas6_grids_forsmall.iloc[(i_n-1)*line_number:i_n*line_number].iterrows():
        sqrt_s = row['sqrt_s']
        x = row['x']
        z = row['z']
        Q = (row['Q'])
        qT_Q = (row['qT_Q'])
        file.write('%f,%f,%f,%f,%f\n'%(sqrt_s,x,z,Q,qT_Q))

```

```python
#for traditional
file = open('clas6/clas6_grids_traditional.txt','w')
file.write('SF_Set_Index,Eb,beam_pol,target_pol,x,y,z,ph_t_sq,phih,phi_s,k0\n')
for i,row in clas6_data.iterrows():
    index = 0
    Eb = beam_energy
    beam_pol ='U'
    target_pol = 'U'
    x = row['x']
    Q2 = (row['Q2'])
    s = 2*Eb*M_P
    y = Q2/(s*x)
    z = row['z']
    ph_t_sq = row['pT2']
    phih=row['phi']
    phis=1
    k0=0.01
    file.write('%i,%f,%s,%s,%f,%f,%f,%f,%f,%f,%f\n'%(index,Eb,beam_pol,target_pol,x,y,z,ph_t_sq,phih,phis,k0))

```
